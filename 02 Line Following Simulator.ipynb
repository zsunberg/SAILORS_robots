```python
from sailorssim import simulate
```

```python
def control(speed, line_position, previous_position):
    k_p = 0.3
    k_d = 0.1
    
    derivative = line_position - previous_position
    delta = k_p * line_position + k_d * derivative
    
    leftspeed = speed + delta
    rightspeed = speed - delta
    
    return leftspeed, rightspeed
```

```python
simulate(control, output='animate') # faster: output='plot'
```
